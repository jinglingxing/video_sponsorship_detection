 ### Word2Vec model

```python
from gensim.models import Word2Vec
import multiprocessing
from time import time 
```

```python
w2v_model = Word2Vec(min_count=3,
                     window=4,
                     vector_size=300,
                     sample=1e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007, 
                     negative=20,
                     workers=multiprocessing.cpu_count()-1)
```

```python
import ast
transcripts = [ast.literal_eval(row) for row in processed_sponsor_df['processed_transcript']]
transcripts[0]
```

```python
from gensim.models.phrases import Phrases, Phraser
phrases = Phrases(transcripts, min_count=1, progress_per=5000)
phrases
```

```python
bigram = Phraser(phrases)
bigram
```

```python
subtitles = bigram[transcripts]
subtitles[0]
```

```python
w2v_model = Word2Vec(min_count=3,
                     window=4,
                     vector_size=300,
                     sample=1e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007, 
                     negative=20,
                     workers=multiprocessing.cpu_count()-1)

start = time()

w2v_model.build_vocab(subtitles, progress_per=50000)

print('Time to build vocab: {} mins'.format(round((time() - start) / 60, 2)))
```

```python
start = time()

w2v_model.train(subtitles, total_examples=w2v_model.corpus_count, epochs=30, report_delay=1)

print('Time to train the model: {} mins'.format(round((time() - start) / 60, 2)))
```

```python
w2v_model.save("word2vec.model")
```

```python
word_vectors = Word2Vec.load('word2vec.model').wv
word_vectors.vectors
```

### KMeans

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=5, max_iter=1000, random_state=True, n_init=50).fit(X=word_vectors.vectors.astype('double'))
```

```python
word_vectors.similar_by_vector(model.cluster_centers_[1], topn=10, restrict_vocab=None)
```

```python
word_vectors.similar_by_vector(model.cluster_centers_[0], topn=10, restrict_vocab=None)
```

```python
word_vectors.similar_by_vector(model.cluster_centers_[4], topn=10, restrict_vocab=None)
```

```python
positive_cluster_index = 1
positive_cluster_center = model.cluster_centers_[positive_cluster_index]
negative_cluster_center = model.cluster_centers_[1-positive_cluster_index]
```

```python
words = pd.DataFrame(word_vectors.index_to_key)
words.columns = ['words']
words
```

```python
words['vectors'] = words.words.apply(lambda x: word_vectors[f'{x}'])
```

```python
import numpy as np
words['cluster'] = words.vectors.apply(lambda x: model.predict([np.array(x)]))
```

```python
words['cluster_value'] = [1 if i==positive_cluster_index else -1 for i in words.cluster]
words['closeness_score'] = words.apply(lambda x: 1/(model.transform([x.vectors]).min()), axis=1)
words['sentiment_coeff'] = words.closeness_score * words.cluster_value
```
